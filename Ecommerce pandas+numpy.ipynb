```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'ecommerce.csv')
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

```python
# 1- What are the total Order value for each country
TotalOrderForEachCountry=df.groupby('Country')['OrderValue'].sum()
TotalOrderForEachCountry.reset_index()
```

```python
# 2- What each Major Category sold quantity during November 2011
N=df[df['Year-Month']=='2011-11']
N.groupby('Major Category').sum()['Quantit'].reset_index()
```

```python
# 3- What's Total Order Value for each Minor Item in Kitchen
TotalOrder=df.groupby(['Major Category','Minor Category'])['OrderValue'].sum()
TotalOrder.reset_index()
```

```python
# 4- What Month/s contributes to high demand in orders
df.groupby('Year-Month')['Quantit'].sum().reset_index()
```

```python
# 5- What Month/s contributes to high OrderValue?
df.groupby('Year-Month')['OrderValue'].sum().reset_index()
```

```python
# 6- Why April and December suffer from low sales? 
Months=['2011-04','2011-12']
AprilAndDec=df[df['Year-Month'].isin(Months)]
AprilAndDec.groupby(['Year-Month','Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

```python
# 7- Why May and November have high sales? 
Month=['2011-05','2011-11']
MayAndNovember=df[df['Year-Month'].isin(Month)]
MayAndNovember.groupby(['Year-Month','Major Category','Minor Category','Country'])['OrderValue'].sum().reset_index()
```

```python
# 8- Most Major Items sold in April and December
AprilAndDec.groupby('Major Category').sum().reset_index()
```

```python
# 9- Most Major Items sold in May and November
MayAndNovember.groupby('Major Category').sum().reset_index()
```

```python
# 10- What Hour recives high demand for ordering?
df['Hour'] = df['InvoiceTime'].str.split(':').str[0]
df.groupby('Hour')['Quantit'].sum().reset_index()
```

```python
# 11- What's the Time of the day that contributes to high Order Values?
def time_of_day(x):
    if x in range(6,12):
        return 'Morning'
    elif x in range(12,16):
        return 'Afternoon'
    elif x in range(16,22):
        return 'Evening'
    else:
        return 'Late night'
df.Hour = pd.to_numeric(df.Hour, errors='coerce')
df['OrderTimeOfDay']=df['Hour'].apply(time_of_day)
df.groupby('OrderTimeOfDay')['Quantit'].sum().reset_index()
```

```python
# 12 - Order VAlue For each Country in each month
pd.pivot_table(df, values='OrderValue', index=['Year-Month'], columns='Country', aggfunc=np.sum).reset_index()
```

```python
# 13 - Order VAlue For each Country and Major Category in Minor Category
pd.pivot_table(df, values='OrderValue', index=['Year-Month'], columns='Country', aggfunc=np.sum).reset_index()
```

```python
# 14 - Order VAlue For each Country and Major Category in Minor Category
pd.pivot_table(df, values='Quantit', index=['Country','Major Category'], columns='Hour', aggfunc='sum').reset_index()
```

```python
# 15- Correlation between 'Quantit','OrderValue','UnitPrice'.
df.corr()[['Quantit','OrderValue','UnitPrice']]
```
